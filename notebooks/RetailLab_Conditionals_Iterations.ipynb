## 3. Merging DataFrames

#### 3.1. Transactions-Articles


```python
Transactions.head(3)
```

```python
Articles.head(3)
```

```python
## we merge both DataFrames using the common key: article_id. We store the result in a new DataFrame
TransactionsAndArticles=pd.merge(Transactions, Articles, how='left',on='article_id')
```

#### 3.2. Transactions-Articles-Customers

```python
## we merge both DataFrames using the common key: customer_id. We store the result in a new DataFrame
TransactionsAndArticlesAndCustomers=pd.merge(TransactionsAndArticles, Customers, how='left',on='customer_id')
```

```python
TransactionsAndArticlesAndCustomers.info()
```

```python
TransactionsAndArticlesAndCustomers.head(2).T
```

```python
TransactionsAndArticlesAndCustomers.groupby('sales_channel_id').count()
```

### 4. Let's explore sales channel's performance

```python
# We rely on Pandas well-known operations to compare the financial performance of each channel
# this approach is always the preferred one

TransactionsAndArticlesAndCustomers.groupby('sales_channel_id')['price'].mean()
```

```python
# A for loop that displays each group's mean
for item,group in TransactionsAndArticlesAndCustomers.groupby(['sales_channel_id']):
  print('I am in group',item[0])
  groupmean=group['price'].mean()
  print("The mean of the group's sales is:",groupmean)
```

```python
for item,group in TransactionsAndArticlesAndCustomers.groupby(['sales_channel_id']):
  print('I am in group',item[0])
  sns.histplot(x=group['price'],bins=10)
  plt.show()
```

### 4. Let's extract items made of cotton (the hard way)
#### 1. From the column `detail_desc` we select rows containing the word "cotton".
#### 2. We create a new column titled: `hasCotton`

```python
def cottonFinder(input):
  #we make sure that the input is a string
  string=str(input)
  if 'cotton' in string:
    return('it has cotton')
  else:
    return('it does not have cotton')
```

```python
cottonFinder('trousers made of cotton')
```

```python
cottonFinder('Short made of microfiber')
```

```python
cottonFinder(3.2)
```

```python
TransactionsAndArticlesAndCustomers['detail_desc']
```

```python
TransactionsAndArticlesAndCustomers['detail_desc'].apply(lambda x:cottonFinder(x))
```

```python
# we store the results in the column `hasCotton`
TransactionsAndArticlesAndCustomers['hasCotton']=TransactionsAndArticlesAndCustomers['detail_desc'].apply(lambda x:cottonFinder(x))
```

```python
TransactionsAndArticlesAndCustomers[['detail_desc','hasCotton']].sample(30)
```

```python
TransactionsAndArticlesAndCustomers.groupby('hasCotton').count()
```

```python
TransactionsAndArticlesAndCustomers[TransactionsAndArticlesAndCustomers['hasCotton']=='it has cotton']
```
