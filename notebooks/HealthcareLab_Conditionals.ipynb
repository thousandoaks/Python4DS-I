## 3. Applying conditionals to a Pandas DataFrame
#### It is advisable to implement conditional-based execution logic only when strictly required. It is better to rely on pandas native operations whenever possible.

```python
HealthCareDataSet['TypeFlag'].apply(lambda x:'Found ER' if x=='ER' else "Did not find ER")
```

```python
HealthCareDataSet['TotalExpenses'].apply(lambda x:'Expense larger than 100' if x>100 else "Expense lower than 100")
```

```python
HealthCareDataSetReduced=HealthCareDataSet.head(1000)
```

```python
## we process each element of the column `PrincipalDiagnosisDesc`

for item in HealthCareDataSetReduced['PrincipalDiagnosisDesc']:
  print(item.upper())
```

```python
## for loops are specially useful to implement fine grained processing of group-by operations
```

```python
HealthCareDataSet.groupby(['TypeFlag']).count()
```

```python
HealthCareDataSet.groupby(['TypeFlag'])['TotalExpenses'].mean()
```

```python
# a for loop that prints the name of each group
for item,group in HealthCareDataSetReduced.groupby(['TypeFlag']):
  print('I am in group',item)
```

```python
# a for loop that displays each group
for item,group in HealthCareDataSetReduced.groupby(['TypeFlag']):
  print('I am in group',item)
  display(group)
```

```python
# A for loop that displays each group's mean
for item,group in HealthCareDataSetReduced.groupby(['TypeFlag']):
  print('I am in group',item)
  groupmean=group['TotalExpenses'].mean()
  print("The mean of the group's Total Expenses is:",groupmean)
```

```python
# A for loop that displays each group's maximum value
for item,group in HealthCareDataSetReduced.groupby(['TypeFlag']):
  print('I am in group',item)
  groupmaxvalue=group['TotalExpenses'].max()
  print("The maxinum value of the group's Total Expenses is:",groupmaxvalue)
```

```python
# A for loop that displays each group's top 3 largest Expenses
for item,group in HealthCareDataSetReduced.groupby(['TypeFlag']):
  print('I am in group',item)
  top3maxvalue=group['TotalExpenses'].nlargest(3)
  print("The maximum value of the group's Total Expenses is:\n",top3maxvalue)
```

```python
# A for loop that plot each group's Total Expenses
for item,group in HealthCareDataSetReduced.groupby(['TypeFlag']):
  print('I am in group',item)
  sns.boxplot(x=group['TotalExpenses'],log_scale=True)
  plt.show()
```
