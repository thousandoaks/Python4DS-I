## Library import

The following line imports the *Seaborn* library.

We rename it sns for convenience purposes

```
import seaborn as sns
```

We also import the `Pandas` library

```
import pandas as pd
```

## Data loading and DataFrame creation


. The following example loads a file with the data. Run the following cell to load the data and create your first `DataFrame`

```
tips = sns.load_dataset("tips")
```

Let's display the first few records of a `DataFrame`:

```
tips.head()
```

## Histograms

In the following figures we represent the empirical distribution of total bill and tip respectively

```
sns.displot(data=tips, x="total_bill", kde=True)
```

```
sns.displot(data=tips, x="tip", kde=True)
```

Histograms are quite effective to discern differences accross groups. Following we observe that dinner time is better in terms of tips

```
sns.displot(data=tips, x="tip", hue="time", kde=True)
```

## ScatterPlots

The following figure reveals the relationship between total_bill and tips. We notice that higher total bills lead to larger tips.

```
sns.lmplot(data=tips, x="total_bill", y="tip")
```

Moreover, as shown in the following figure, we observe that the total_bill tip relationship is contingent on being a smoker. Non-smokers seem to be more generous than their smoking counterparts

```
sns.lmplot(data=tips, x="total_bill", y="tip" ,hue="smoker")
```
