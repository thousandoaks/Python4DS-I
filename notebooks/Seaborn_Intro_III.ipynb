**Learning Objectives:**

## Library import

The following line imports the *Seaborn* library.

We rename it sns for convenience purposes

```
import seaborn as sns
```

We also import the `Pandas` library

```
import pandas as pd
```

## Data loading and DataFrame creation


. The following example loads a file with the  data. Run the following cell to load the data and create your first `DataFrame`

```
tips = sns.load_dataset("tips")
```

Let's display the first few records of a `DataFrame`:

```
tips.head()
```

## BarPlot

```
sns.barplot(x="time", y="total_bill", data=tips)
```

```
sns.barplot(x="day", y="total_bill", data=tips)
```

```
sns.barplot(x="day", y="total_bill", hue='time',data=tips)
```

## BoxPlot

```
sns.boxplot(x="size", y="tip", data=tips)
```

```
sns.boxplot(x="sex", y="tip", data=tips)
```

```
sns.boxplot(x="day", y="tip", hue='smoker',data=tips)
```

## ViolinPlot

ViolingPlots are very similar to BoxPlots

```
sns.violinplot(x="size", y="tip", data=tips)
```

```
sns.violinplot(x="day", y="tip", hue='smoker',data=tips)
```

## PairPlot

```
sns.pairplot(tips, hue="smoker")
```
